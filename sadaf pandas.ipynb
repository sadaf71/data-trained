```python
#merge
rto_offc= pd.DataFrame({'adhaar':['a123','b234','c345','d456','e567','f678','g789','h890'],
                        'vechile_details':['honda','chery','suzuki','hyundai','yamaha','mahindra','honda','gm ford'],
                        'license':['yes','yes','no','yes','no','yes','yes','yes']})
rto_offc
```

```python
bajaj_off_insu= pd.DataFrame({'adhaar':['a123','b234','e567','h890'],
                              'accident':['yes','yes','yes','yes'],
                              'claim_amt':[450000,54000,95000,420000]})
bajaj_off_insu
```

```python
final_merge= pd.merge(rto_offc,bajaj_off_insu,on='adhaar')
final_merge
```

```python
claim_accepted=[]
for claim in final_merge['license']:
    if claim == 'yes':
        claim_accepted.append('accepted')
    else:
        claim_accepted.append('rejected')
            
final_merge['claim_accepted?']= claim_accepted
final_merge
```
